```python
df=pd.DataFrame(i.data,columns=i.feature_names)
df["species"]=i.target
```

```python
df.head()
```

```python
setosa=df[df["species"]==0]
versi=df[df["species"]==1]
vergi=df[df["species"]==2]
setosa.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(setosa["sepal length (cm)"],setosa["sepal width (cm)"],label="setosa")
plt.scatter(versi["sepal length (cm)"],versi["sepal width (cm)"],color ="red",label="versi" )
plt.scatter(vergi["sepal length (cm)"],vergi["sepal width (cm)"],color="magenta",label="vergi")
plt.legend()
```

```python
plt.scatter(setosa["petal length (cm)"],setosa["petal width (cm)"],label="setosa")
plt.scatter(versi["petal length (cm)"],versi["petal width (cm)"],color ="red",label="versi" )
plt.scatter(vergi["petal length (cm)"],vergi["petal width (cm)"],color="magenta",label="vergi")
plt.legend()
```

```python
df
```

```python
x=df.drop("species",axis=1)
x
```

```python
y=df.species
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
len(train_x),len(train_y),len(test_x),len(test_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(train_x,train_y)
```

```python

pred = model.predict([[5, 8, 5, 6]])

if pred[0] == 0:
    print("setosa")
elif pred[0] == 1:
    print("versicolor")
else:
    print("virginica")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model.score(test_x,test_y)
```

```python
for i in range(1,50):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(train_x,train_y)
    print("value of i is ",i,"      ",model.score(test_x,test_y))
```
